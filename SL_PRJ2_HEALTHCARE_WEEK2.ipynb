1. Check the balance of the data by plotting the count of outcomes by their value. Describe your findings and plan future course of action.

```python

sns.set_style('darkgrid')
sns.countplot(data['Outcome'])
plt.title("Countplot of Outcome")
plt.xlabel('Outcome')
plt.ylabel("Count")
```

We can see that value (1 - Diabetes) is more than 50% of value 0(non-diabetes). SO data is balanced. I am taking into account - 30% as the relative standard, if one of the data is less than 30% of the other data, then we can consider data as imbalanced and we need to consider sampling techniques for better model accuracy. But here we can take the data As it is for model building and classification alogirthms.
